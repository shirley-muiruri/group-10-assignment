##PASSING IN A FUNCTION


```python
#re-reading the data
df = pd.read_csv('turnstile_180901.txt')
df
```

```python
#accessing the LINENAME column
df['LINENAME'].value_counts()
```

```python
# define a function that takes in a string
# returns True if that string contains the letter 'N', and returns False otherwise
def contains_n(LINENAME):
    if 'N' in LINENAME:
        return True
    else:
        return False
        contains_n("LINENAME")
```

```python
#shorter code
def contains_n(LINENAME):
    return 'N' in LINENAME
    contains_n("LINENAME")
```

```python
#using .map() pass in the function
df['LINENAME'] = df['LINENAME'].map(contains_n)
df['LINENAME']
```

```python
#using .apply()method
df['LINENAME'].apply(contains_n)
```

```python
#Creating a new column
#representing the results
df['On_N_Line'] = df['LINENAME'].map(contains_n)
df
```

```python
df['On_N_Line'].value_counts(normalize=True)
```

```python
#using a lambda function
#lambda is an anonymous function
# x represents each value in the LINENAME column
df['On_N_Line'] = df['LINENAME'].map(lambda x: 'N' in x)
df
```

```python
#checking whether a string contains a string
df['On_N_Line']= df['LINENAME'].str.contains('N', regex=False)
df['On_N_Line']
```
